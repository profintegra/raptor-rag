1.1 Đặt vấn đề
Trí tuệ nhân tạo (AI, Artificial Intelligence) đang ngày càng được ứng dụng rộng
rãi trong nhiều lĩnh vực của đời sống, trở thành công cụ quan trọng giúp con người
nâng cao năng suất làm việc và cải thiện chất lượng cuộc sống. Trong giai đoạn
chuyển đổi số như hiện nay, Trí tuệ nhân tạo đóng vai trò then chốt, thúc đẩy sự
đổi mới và tối ưu hóa các hoạt động kinh doanh, quản lý và vận hành. Một minh
chứng rõ nét cho sự phát triển vượt bậc của nó chính là sự nổi lên mạnh mẽ của các
mô hình ngôn ngữ lớn, tích hợp trong các hệ thống hỏi đáp tự động như ChatGPT,
Gemini, . . . với khả năng xử lý và tạo ngôn ngữ tự nhiên, cho phép máy tính tương
tác và hỗ trợ con người một cách hiệu quả hơn bao giờ hết.
Với kích thước vô cùng lớn và được huấn luyện trên khối lượng dữ liệu khổng lồ,
các mô hình ngôn ngữ lớn có khả năng học được cấu trúc ngôn ngữ phức tạp, nắm
bắt ngữ nghĩa và tích lũy một lượng lớn tri thức thế giới (world knowledge) từ các
nguồn dữ liệu mà chúng đã được “đọc” /ref. Điều này giúp chúng có thể thực hiện
được đa dạng các tác vụ ngôn ngữ như tạo văn bản, tóm tắt thông tin, dịch thuật,
trả lời câu hỏi. . . một cách tự nhiên và mạch lạc /ref. Sự phổ biến của các công cụ,
như ChatGPT, cho thấy các mô hình ngôn ngữ lớn ngày càng được sử dụng rộng
rãi để trả lời các câu hỏi thường ngày. Vì thế, độ chính xác của thông tin mà các
mô hình này cung cấp ngày càng trở nên quan trọng. Tuy nhiên, các mô hình ngôn
ngữ lớn thường gặp hiện tượng "ảo giác" (hallucination), nói cách khác, câu trả lời
mà chúng tạo ra có vẻ trôi chảy và mạch lạc, nhưng lại chứa thông tin không chính
xác nên rất khó để phát hiện được những thông tin sai lệch này. /ref Hình 1.1 minh
họa một ví dụ về hiện tượng ảo giác của các mô hình ngôn ngữ lớn. Trong ví dụ
này, mô hình dù đưa ra lời giải thích có vẻ hợp lý "tên bốn người là bốn mùa trong
năm" nhưng lại không đúng (vì người thứ tư tên là Nam).
Hiện tượng ảo giác hình thành trong suốt quá trình tiếp thu tri thức của các mô
hình ngôn ngữ lớn, thường được bắt nguồn từ ba vấn đề chính: dữ liệu, quá trình
huấn luyện và suy luận/ref. Một số vấn đề về dữ liệu có thể kể đến bao gồm: i)
thông tin sai lệch và thành kiến trong nguồn dữ liệu huấn luyện, ii) chất lượng kém,
dữ liệu trùng lặp làm ảnh hưởng đến hiệu năng mô hình, và đặc biệt là iii) giới hạn
về kiến thức khi tri thức của các mô hình ngôn ngữ lớn dừng lại tại thời điểm huấn

```
Cảm ơn bạn đã cung cấp thông tin rất chi tiết về RAPTOR. Dưới đây là một tóm tắt đơn giản và gọn gàng về RAPTOR:

---

### **RAPTOR (Recursive Abstractive Processing for Tree-Organized Retrieval)**

#### **Tổng quan:**
RAPTOR là một phương pháp tiên tiến trong lĩnh vực **retrieval-augmented generation (RAG)**, được thiết kế để vượt qua những hạn chế của các kỹ thuật truy xuất thông tin truyền thống. Điểm độc đáo của RAPTOR là nó sử dụng một cấu trúc **cây phân cấp** để tổ chức thông tin theo nhiều mức độ trừu tượng. Cách tiếp cận này không chỉ giúp hiểu sâu sắc hơn các ngữ cảnh lớn mà còn hỗ trợ tích hợp thông tin chi tiết từ các đoạn văn bản dài.

#### **Phương pháp:**
1. **Xây dựng cây:**
   - Văn bản được chia thành các đoạn nhỏ (chunks), sau đó được nhúng (embedding) bằng các mô hình như SBERT.
   - Các đoạn nhúng này được nhóm lại bằng thuật toán **Gaussian Mixture Models (GMM)** và tóm tắt thành các nút cha. Quy trình này lặp lại theo cách đệ quy để tạo thành cây với nhiều cấp độ.

2. **Tóm tắt đệ quy:**
   - Mỗi cụm văn bản (cluster) được tóm tắt lại bằng các mô hình ngôn ngữ lớn (LLM). Tóm tắt này đại diện cho nội dung của các nút con, tạo nên các mức độ trừu tượng khác nhau.

3. **Truy vấn trong cây:**
   - Có hai phương pháp truy vấn:
     - **Tree Traversal:** Truy vấn từ gốc cây và đi qua các nút con đến lá.
     - **Collapsed Tree:** Làm phẳng toàn bộ cây thành một tầng và truy vấn toàn bộ các nút.

#### **Ưu điểm và Nhược điểm:**
- **Ưu điểm:**
  - **Hiểu ngữ cảnh sâu sắc:** RAPTOR hỗ trợ lấy thông tin từ nhiều mức độ trừu tượng, phù hợp với các bài toán phức tạp đòi hỏi tư duy nhiều bước.
  - **Hiệu quả với dữ liệu dài:** Các cấp độ trừu tượng giúp tích hợp thông tin từ nhiều phần của văn bản dài, vượt trội hơn so với cách lấy thông tin theo đoạn ngắn liên tiếp.
  - **Hiệu suất cao:** RAPTOR đạt kết quả tốt hơn trong các bài kiểm tra như NarrativeQA, QASPER, và QuALITY, với độ chính xác vượt qua các mô hình truy xuất truyền thống như DPR và BM25.

- **Nhược điểm:**
  - **Chi phí tính toán cao:** Việc xây dựng cây và tóm tắt đệ quy đòi hỏi nhiều tài nguyên tính toán.
  - **Phức tạp trong triển khai:** Yêu cầu tích hợp các thuật toán nhóm, nhúng, và tóm tắt, đòi hỏi sự am hiểu sâu về NLP và hệ thống.

#### **Kết quả:**
- RAPTOR đã chứng minh hiệu quả vượt trội trong các bài toán hỏi đáp phức tạp, ví dụ:
  - Tăng **20% độ chính xác tuyệt đối** trên bộ dữ liệu QuALITY khi kết hợp với GPT-4.
  - Cải thiện hiệu suất trên các bài kiểm tra dài (NarrativeQA) nhờ khả năng tóm tắt và truy vấn đa cấp.

#### **Kết luận:**
RAPTOR là một bước tiến lớn trong việc tăng cường hiệu suất của các mô hình ngôn ngữ bằng cách tổ chức và truy xuất thông tin thông minh. Phương pháp này đặc biệt phù hợp cho các ứng dụng yêu cầu tích hợp thông tin phức tạp, như hỏi đáp đa bước, phân tích ngữ cảnh chuyên sâu, và nghiên cứu dữ liệu lớn.

---

Nếu bạn cần triển khai RAPTOR trong dự án cụ thể hoặc muốn viết phần này bằng LaTeX, cứ yêu cầu mình nhé! 😊
```
