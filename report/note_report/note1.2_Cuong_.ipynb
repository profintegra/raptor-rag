```
\section{Các phương pháp Chunking truyền thống}

\begin{table}[h]
    \centering
    \caption{Bảng so sánh các chiến lược Chunking (Cập nhật)}
    \label{tab:chunking_methods}
    \begin{tabular}{|p{5cm}|p{6cm}|p{5cm}|p{5cm}|}
        \hline
        \textbf{Chiến lược Chunking} & \textbf{Mô tả} & \textbf{Ưu điểm} & \textbf{Nhược điểm} \\ \hline
        A. Fixed-size Chunking & Chia văn bản thành các đoạn có độ dài cố định với độ overlap. & Dễ triển khai, các chunk có độ dài đồng đều. & Dễ cắt ngang câu và đoạn văn, có thể làm mất ngữ cảnh quan trọng. \\ \hline
        B. Chunking Theo Câu & Chia văn bản dựa trên các câu hoàn chỉnh, giữ nguyên ngữ cảnh. & Bảo toàn ngữ cảnh, phù hợp với nội dung có cấu trúc phức tạp. & Các chunk có độ dài không đồng đều, có thể dẫn đến chunk quá dài. \\ \hline
        C. Chunking Kết Hợp (Hybrid) & Kết hợp giữa Fixed-size và Sentence-based, ưu tiên chia tại đầu câu. & Giữ ngữ cảnh tốt hơn Fixed-size, đặc biệt hữu dụng cho sentence-based. & Phức tạp hơn để triển khai. \\ \hline
        D. Chunk theo Ngữ Nghĩa & Sử dụng mô hình ngôn ngữ để chia các đoạn theo ngữ nghĩa thay vì chiều dài cố định. & Tối ưu cho tìm kiếm ngữ nghĩa, giúp bảo toàn ý nghĩa tốt hơn. & Yêu cầu tài nguyên tính toán lớn, khó triển khai. \\ \hline
        E. Tách Đệ Quy Ký Tự & Chia nhỏ đệ quy dựa trên các dấu phân cách như dấu chấm câu. & Giữ cấu trúc tốt hơn tách ký tự đơn giản, phù hợp cho văn bản có cấu trúc rõ ràng. & Có thể cắt ngang ý nghĩa nếu dấu phân cách không hợp lý. \\ \hline
        F. Agentic Chunker (Tách Dùng Agent) & Sử dụng các "agent" để tự động nhóm các propositions liên quan thành chunks dựa trên ngữ nghĩa. & Tự động tối ưu theo ngữ cảnh, linh hoạt và thông minh hơn. & Phức tạp, đòi hỏi mô hình AI nâng cao. \\ \hline
        I. Tách Theo Loại Tài Liệu & Phân tách dựa trên loại tài liệu (PDF, Markdown, Python code). & Tối ưu cho các loại tài liệu cụ thể, giúp bảo toàn ngữ nghĩa. & Cần phân loại tài liệu trước khi chia nhỏ. \\ \hline
    \end{tabular}
\end{table}

```
