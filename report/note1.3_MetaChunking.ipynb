```
\section{Phương pháp Meta Chunking}

\subsection{Tổng quan}
Phương pháp \textbf{Meta Chunking} được đề xuất nhằm cải thiện việc phân đoạn văn bản bằng cách tạo ra các khối nội dung (\textit{meta-chunk}) với tính liên kết logic cao. Không giống như các phương pháp phân đoạn truyền thống dựa trên câu hoặc đoạn văn, Meta Chunking hoạt động ở cấp độ chi tiết hơn, trong đó mỗi khối nội dung bao gồm một tập hợp các câu có mối quan hệ ngữ nghĩa và logic sâu sắc, như quan hệ nhân quả, chuyển tiếp, song song, hoặc tuần tự.

\subsection{Cơ sở lý thuyết}
Phương pháp Meta Chunking dựa trên hai trụ cột chính:
\begin{itemize}
    \item \textbf{Đo lường độ phức tạp (Perplexity - PPL):} Sử dụng độ phức tạp ngữ cảnh để xác định các điểm ngắt hợp lý trong văn bản, đánh dấu ranh giới giữa các khối nội dung. Điểm ngắt được xác định tại những vị trí mà độ phức tạp thay đổi đột ngột.
    \item \textbf{Kết hợp động (Dynamic Combination):} Sau khi tạo ra các khối nội dung nhỏ (\textit{meta-chunk}), các khối này được hợp nhất lại để đạt được độ dài mong muốn, cân bằng giữa chi tiết ngữ nghĩa và hiệu quả xử lý.
\end{itemize}

\subsection{Quy trình thực hiện}
Phương pháp Meta Chunking bao gồm ba bước chính như sau:

\subsubsection{Bước 1: Tách câu và tính độ phức tạp}
Văn bản được chia thành các câu \((x_1, x_2, \ldots, x_n)\). Độ phức tạp của mỗi câu \(PPL_M(x_i)\) được tính như sau:
\[
PPL_M(x_i) = \frac{1}{K} \sum_{k=1}^K PPL_M(t_k \mid t_{<k}, t_{<i})
\]
Trong đó:
\begin{itemize}
    \item \(t_k\): Token thứ \(k\) trong câu \(x_i\).
    \item \(t_{<k}\): Các token trước đó trong câu \(x_i\).
    \item \(t_{<i}\): Các token trong các câu trước \(x_i\).
\end{itemize}

\subsubsection{Bước 2: Xác định ranh giới khối nội dung}
Phân bố độ phức tạp \(PPL_{\text{seq}} = (PPL_M(x_1), PPL_M(x_2), \ldots, PPL_M(x_n))\) được phân tích để tìm các điểm tối thiểu cục bộ \(PPL_{\text{min}}\):
\[
\text{MinimaIndex}(PPL_{\text{seq}}) = \{i \mid PPL_M(x_{i-1}) - PPL_M(x_i) > \theta \text{ hoặc } PPL_M(x_{i+1}) - PPL_M(x_i) > \theta\}
\]
Trong đó:
\begin{itemize}
    \item \(\theta\): Ngưỡng để xác định ranh giới giữa các khối nội dung dựa trên sự thay đổi đột ngột của độ phức tạp.
    \item Các điểm có \(PPL_M(x_i)\) nhỏ hơn \(PPL_M(x_{i-1})\) và \(PPL_M(x_{i+1})\) được xem là ranh giới khối nội dung.
\end{itemize}

\subsubsection{Bước 3: Kết hợp động (Dynamic Combination)}
\begin{enumerate}
    \item Sau khi xác định ranh giới, các khối nội dung nhỏ được tạo ra. Để đạt được độ dài khối mong muốn, các khối này được hợp nhất theo thuật toán kết hợp động.
    \item Các khối được hợp nhất nếu tổng độ dài của chúng \(\text{len}(c_1, c_2, c_3)\) nhỏ hơn ngưỡng \(L\):
    \[
    \begin{aligned}
    \text{Nếu } \text{len}(c_1, c_2, c_3) &< L, \text{ hợp nhất } c_1, c_2, c_3. \\
    \text{Ngược lại, giữ nguyên cấu trúc.}
    \end{aligned}
    \]
\end{enumerate}

\subsection{Thực nghiệm}
Phương pháp Meta Chunking được kiểm tra trên 11 tập dữ liệu, bao gồm các nhiệm vụ hỏi đáp (QA) đơn cấp và đa cấp. Kết quả cho thấy:
\begin{itemize}
    \item \textbf{Hiệu suất cao:} Meta Chunking cải thiện đáng kể độ chính xác câu trả lời so với các phương pháp dựa trên quy tắc hoặc tương đồng ngữ nghĩa.
    \item \textbf{Tính hiệu quả:} Giảm thời gian xử lý xuống còn khoảng 45\% so với các phương pháp khác mà vẫn đảm bảo chất lượng.
    \item \textbf{Tính linh hoạt:} Phương pháp hoạt động tốt trên cả mô hình nhỏ (<1B tham số) và mô hình lớn (>7B tham số).
\end{itemize}

\subsection{Ứng dụng}
Phương pháp Meta Chunking có thể được ứng dụng trong nhiều lĩnh vực, bao gồm:
\begin{itemize}
    \item \textbf{Hệ thống hỏi đáp (QA):} Tăng cường khả năng truy xuất và sinh câu trả lời trong các nhiệm vụ yêu cầu kiến thức chuyên sâu.
    \item \textbf{Xử lý văn bản dài:} Tối ưu hóa việc xử lý tài liệu lớn, giảm thiểu thông tin dư thừa.
    \item \textbf{Tóm tắt tài liệu:} Tạo ra các bản tóm tắt ngắn gọn nhưng đầy đủ thông tin.
\end{itemize}

\subsection{Kết luận}
Phương pháp Meta Chunking mang lại một cách tiếp cận mới trong phân đoạn văn bản, tập trung vào tính mạch lạc và hiệu quả xử lý. Với khả năng tổ chức thông tin một cách hợp lý, Meta Chunking không chỉ cải thiện hiệu suất của các hệ thống mà còn tiết kiệm tài nguyên và thời gian xử lý.


```

```
\section{Meta Chunking Method}

\subsection{Overview}
The \textbf{Meta Chunking} method is proposed to enhance text segmentation by creating content blocks (\textit{meta-chunks}) with high logical coherence. Unlike traditional segmentation methods based on sentences or paragraphs, Meta Chunking operates at a finer granularity, where each content block consists of a set of sentences with deep semantic and logical relationships, such as causal, transitional, parallel, or sequential connections.

\subsection{Theoretical Foundation}
The Meta Chunking method is based on two main pillars:
\begin{itemize}
    \item \textbf{Perplexity Measurement (PPL):} Contextual perplexity is used to identify appropriate breakpoints in the text, marking the boundaries between content blocks. Breakpoints are determined at positions where perplexity changes abruptly.
    \item \textbf{Dynamic Combination:} After creating small content blocks (\textit{meta-chunks}), these blocks are merged to achieve the desired length, balancing semantic detail and processing efficiency.
\end{itemize}

\subsection{Implementation Process}
The Meta Chunking method consists of three main steps as follows:

\subsubsection{Step 1: Sentence Splitting and Perplexity Calculation}
The text is divided into sentences \((x_1, x_2, \ldots, x_n)\). The perplexity of each sentence \(PPL_M(x_i)\) is calculated as follows:
\[
PPL_M(x_i) = \frac{1}{K} \sum_{k=1}^K PPL_M(t_k \mid t_{<k}, t_{<i})
\]
Where:
\begin{itemize}
    \item \(t_k\): The \(k\)-th token in sentence \(x_i\).
    \item \(t_{<k}\): Tokens preceding \(t_k\) in sentence \(x_i\).
    \item \(t_{<i}\): Tokens in sentences preceding \(x_i\).
\end{itemize}

\subsubsection{Step 2: Identifying Content Block Boundaries}
The perplexity distribution \(PPL_{\text{seq}} = (PPL_M(x_1), PPL_M(x_2), \ldots, PPL_M(x_n))\) is analyzed to find local minima \(PPL_{\text{min}}\):
\[
\text{MinimaIndex}(PPL_{\text{seq}}) = \{i \mid PPL_M(x_{i-1}) - PPL_M(x_i) > \theta \text{ or } PPL_M(x_{i+1}) - PPL_M(x_i) > \theta\}
\]
Where:
\begin{itemize}
    \item \(\theta\): Threshold to determine boundaries between content blocks based on abrupt changes in perplexity.
    \item Points where \(PPL_M(x_i)\) is smaller than \(PPL_M(x_{i-1})\) and \(PPL_M(x_{i+1})\) are considered content block boundaries.
\end{itemize}

\subsubsection{Step 3: Dynamic Combination}
\begin{enumerate}
    \item After identifying the boundaries, small content blocks are created. To achieve the desired block length, these blocks are merged using a dynamic combination algorithm.
    \item Blocks are merged if their total length \(\text{len}(c_1, c_2, c_3)\) is smaller than the threshold \(L\):
    \[
    \begin{aligned}
    \text{If } \text{len}(c_1, c_2, c_3) &< L, \text{ merge } c_1, c_2, c_3. \\
    \text{Otherwise, retain the structure.}
    \end{aligned}
    \]
\end{enumerate}

\subsection{Experiments}
The Meta Chunking method was tested on 11 datasets, including single-level and multi-level question-answering (QA) tasks. The results show:
\begin{itemize}
    \item \textbf{High Performance:} Meta Chunking significantly improves answer accuracy compared to rule-based or semantic similarity-based methods.
    \item \textbf{Efficiency:} Processing time is reduced by approximately 45\% compared to other methods while maintaining quality.
    \item \textbf{Flexibility:} The method performs well on both small models (<1B parameters) and large models (>7B parameters).
\end{itemize}

\subsection{Applications}
The Meta Chunking method can be applied in various fields, including:
\begin{itemize}
    \item \textbf{Question-Answering Systems (QA):} Enhancing retrieval and answer generation capabilities in tasks requiring in-depth knowledge.
    \item \textbf{Long Text Processing:} Optimizing the handling of large documents while minimizing redundant information.
    \item \textbf{Document Summarization:} Creating concise yet comprehensive summaries.
\end{itemize}

\subsection{Conclusion}
The Meta Chunking method offers a novel approach to text segmentation, focusing on coherence and processing efficiency. By organizing information logically, Meta Chunking not only improves the performance of systems but also saves resources and processing time.

```
