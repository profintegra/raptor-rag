```latex
\documentclass[../DoAn.tex]{subfiles}
\begin{document}

\section{Phương pháp RAPTOR}

\begin{figure}[h]
    \centering
    \includegraphics[width=0.9\textwidth]{Hinh_ve/base_RAPTOR_RAG.png}
    \caption{RAPTOR (Recursive Abstractive Processing for Tree-Organized Retrieval)}
    \label{fig:raptor_overview}
\end{figure}

\subsection{Tổng quan phương pháp}

RAPTOR (Recursive Abstractive Processing for Tree-Organized Retrieval) là một phương pháp tiên tiến trong lĩnh vực truy xuất thông tin hỗ trợ sinh văn bản (Retrieval-Augmented Generation - RAG). Phương pháp này kết hợp giữa kỹ thuật truy xuất thông tin và tạo văn bản bằng cách xây dựng một cấu trúc cây phân cấp đệ quy từ dưới lên. 

Quá trình xây dựng cây RAPTOR bắt đầu bằng việc chia nhỏ các tài liệu đầu vào thành các đoạn văn bản ngắn hơn, mỗi đoạn chứa khoảng 100 từ để đảm bảo tính đồng nhất và ngữ cảnh hợp lý. Các đoạn văn bản này sau đó được chuyển đổi thành các vector nhúng ngữ nghĩa thông qua mô hình SBERT (Sentence-BERT), cho phép biểu diễn các mối quan hệ ngữ nghĩa giữa các đoạn văn bản.

Tiếp theo, các vector nhúng này được nhóm lại thành các cụm dựa trên độ tương đồng ngữ nghĩa bằng thuật toán Gaussian Mixture Models (GMM), đảm bảo rằng các đoạn văn bản trong cùng một cụm có ý nghĩa tương tự nhau. Mỗi cụm sau đó được tóm tắt thành một nút cha bằng cách sử dụng mô hình ngôn ngữ GPT-3.5-turbo, tạo thành các lớp trừu tượng hóa khác nhau trong cây. Quá trình này lặp lại đệ quy cho đến khi không thể phân cụm thêm, tạo nên một cây tổ chức thông tin với nhiều mức độ trừu tượng từ chi tiết đến tổng quan (Nguồn: \href{https://arxiv.org/html/2401.18059v1}{arXiv 2401.18059v1}).

\subsection{Cấu trúc và Xây dựng Cây RAPTOR}

\subsubsection{Chia nhỏ và Nhúng Vector}

Quá trình xây dựng cây RAPTOR bắt đầu bằng việc chia nhỏ các tài liệu thành các đoạn văn bản nhỏ hơn, mỗi đoạn khoảng 100 từ. Việc chia nhỏ này không chỉ giúp duy trì tính đồng nhất của ngữ cảnh mà còn tối ưu hóa quá trình nhúng và phân cụm sau này. Các đoạn văn bản này được nhúng vào không gian vector sử dụng SBERT (Reimers \& Gurevych, 2019), cho phép mô hình hiểu và biểu diễn các mối quan hệ ngữ nghĩa giữa các đoạn văn bản.

\subsubsection{Phân Cụm Ngữ Nghĩa}

Sau khi có các vector nhúng, RAPTOR sử dụng Gaussian Mixture Models (GMM) để phân cụm các đoạn văn bản dựa trên độ tương đồng ngữ nghĩa. GMM cho phép mỗi đoạn văn bản có thể thuộc về nhiều cụm khác nhau, giúp mô hình linh hoạt hơn trong việc nhóm các đoạn văn bản liên quan đến nhiều chủ đề khác nhau. Để tăng cường hiệu quả phân cụm trong không gian vector cao chiều, RAPTOR áp dụng kỹ thuật giảm chiều bằng Uniform Manifold Approximation and Projection (UMAP), giúp duy trì cấu trúc ngữ nghĩa đồng thời giảm thiểu chi phí tính toán.

\subsubsection{Tóm tắt và Xây dựng Cấu Trúc Cây}

Sau khi phân cụm, mỗi cụm sẽ được tóm tắt thành một nút cha bằng cách sử dụng GPT-3.5-turbo. Quá trình tóm tắt này giúp giảm kích thước dữ liệu cần xử lý mà vẫn giữ lại các thông tin cốt lõi. Các nút cha này sau đó trở thành các nút con cho bước phân cụm tiếp theo, và quá trình này được lặp lại cho đến khi không thể phân cụm thêm. Kết quả là một cấu trúc cây phân cấp, trong đó mỗi lớp đại diện cho một mức độ trừu tượng khác nhau của thông tin từ chi tiết đến tổng quan.

\subsection{Chiến lược Truy vấn trong Cây RAPTOR}

RAPTOR hỗ trợ hai chiến lược truy vấn chính: \textbf{Truyền cây theo lớp} (Tree Traversal) và \textbf{Truy vấn cây gộp} (Collapsed Tree Retrieval).

\subsubsection{Truyền cây theo lớp (Tree Traversal)}

Phương pháp này bắt đầu từ lớp gốc của cây và tiến hành truy cập các nút con dựa trên độ tương đồng với truy vấn. Tại mỗi lớp, RAPTOR chọn top-$k$ nút có độ tương đồng cao nhất và tiếp tục truy cập vào các nút con của chúng. Quá trình này lặp lại cho đến khi đạt đến lớp lá. Cuối cùng, tất cả các thông tin từ các nút đã chọn tại các lớp khác nhau được gộp lại để tạo thành ngữ cảnh truy vấn cho mô hình ngôn ngữ lớn (LLM).

\subsubsection{Truy vấn cây gộp (Collapsed Tree Retrieval)}

Phương pháp này gộp toàn bộ các lớp của cây RAPTOR thành một lớp duy nhất và thực hiện tìm kiếm độ tương đồng trực tiếp trên tất cả các nút. RAPTOR sau đó chọn các nút có độ tương đồng cao nhất cho đến khi đạt được giới hạn số lượng token đã định. Phương pháp này linh hoạt hơn vì nó cho phép chọn các nút ở bất kỳ mức độ trừu tượng nào phù hợp với truy vấn, nhưng đòi hỏi phải sử dụng các thư viện tìm kiếm hiệu quả như FAISS để giảm thời gian tính toán.

\subsection{Ưu điểm và Nhược điểm}

\subsubsection*{Ưu điểm}

\begin{enumerate}
    \item \textbf{Tăng cường khả năng truy xuất thông tin:} Cấu trúc cây phân cấp cho phép RAPTOR truy cập thông tin ở nhiều mức độ chi tiết khác nhau, từ tổng quan đến chi tiết, giúp cải thiện khả năng hiểu và trả lời các câu hỏi phức tạp.
    
    \item \textbf{Hiệu suất cao trong các bài toán phức tạp:} Các thử nghiệm cho thấy RAPTOR cải thiện đáng kể độ chính xác trên các tập dữ liệu yêu cầu suy luận đa bước, như QuALITY và NarrativeQA, so với các phương pháp truy xuất truyền thống.
    
    \item \textbf{Khả năng mở rộng linh hoạt:} Với cấu trúc phân cấp, RAPTOR có thể xử lý hiệu quả các tập dữ liệu lớn mà không làm giảm độ chính xác, phù hợp cho các ứng dụng như tóm tắt tài liệu dài và hỏi đáp chuyên sâu.
    
    \item \textbf{Khắc phục hạn chế của mô hình ngôn ngữ lớn:} Bằng cách cung cấp thông tin từ các tài liệu bên ngoài, RAPTOR giảm thiểu hiện tượng "ảo giác" (hallucination) của các mô hình ngôn ngữ lớn, tăng độ tin cậy và chính xác của câu trả lời.
\end{enumerate}

\subsubsection*{Nhược điểm}

\begin{enumerate}
    \item \textbf{Chi phí tính toán cao:} Việc xây dựng và duy trì cấu trúc cây đệ quy đòi hỏi tài nguyên tính toán lớn, từ việc nhúng vector đến phân cụm và tóm tắt, điều này có thể làm tăng chi phí vận hành.
    
    \item \textbf{Độ phức tạp triển khai:} RAPTOR yêu cầu kiến thức chuyên sâu về xử lý ngôn ngữ tự nhiên, học máy, và tổ chức dữ liệu, đòi hỏi đội ngũ kỹ thuật giàu kinh nghiệm để triển khai hiệu quả.
    
    \item \textbf{Độ trễ khi xử lý thời gian thực:} Quá trình tạo cấu trúc cây và tóm tắt đòi hỏi thời gian xử lý đáng kể, khiến RAPTOR không phù hợp với các ứng dụng yêu cầu phản hồi tức thời.
\end{enumerate}

\subsection{Ứng dụng của RAPTOR}

RAPTOR được áp dụng rộng rãi trong nhiều lĩnh vực nhờ khả năng tổ chức và xử lý thông tin vượt trội:

\subsubsection*{Hệ thống Hỏi đáp (QA)}

RAPTOR cung cấp các câu trả lời chính xác và chi tiết bằng cách tổ chức và tích hợp thông tin từ nhiều mức độ ngữ cảnh. Đây là giải pháp lý tưởng cho các ứng dụng hỏi đáp trong giáo dục, tài chính, và y tế, nơi mà các câu hỏi thường đòi hỏi sự hiểu biết sâu sắc và tổng hợp thông tin từ nhiều nguồn khác nhau.

\subsubsection*{Tóm tắt Tài liệu}

RAPTOR tạo ra các bản tóm tắt ngắn gọn nhưng toàn diện từ các tài liệu dài, giúp người dùng tiết kiệm thời gian và tập trung vào các thông tin quan trọng. Điều này đặc biệt hữu ích trong các lĩnh vực như nghiên cứu khoa học, pháp lý, và báo chí, nơi mà việc xử lý và tóm tắt thông tin lớn là cần thiết.

\subsubsection*{Hỗ trợ Nghiên cứu}

Với khả năng phân tích và tổ chức thông tin, RAPTOR hỗ trợ điều hướng và hiểu các tài liệu học thuật phức tạp, đặc biệt trong các lĩnh vực như khoa học đời sống và kỹ thuật. RAPTOR giúp các nhà nghiên cứu nhanh chóng tìm kiếm và tổng hợp các thông tin liên quan, tăng cường hiệu quả nghiên cứu.

\subsubsection*{Trợ lý Ảo Thông minh}

Bằng cách cung cấp câu trả lời chính xác và ngữ cảnh sâu sắc, RAPTOR cải thiện đáng kể trải nghiệm người dùng trong các trợ lý ảo, từ hỗ trợ khách hàng đến các hệ thống tự động hóa doanh nghiệp. Các trợ lý ảo sử dụng RAPTOR có thể hiểu và phản hồi các yêu cầu phức tạp một cách hiệu quả hơn, tăng cường tính hữu ích và độ tin cậy.

\subsection{Quá trình Xây dựng và Tối ưu hóa Cấu trúc Cây}

\subsubsection{Embedding và Phân Cụm}

Quá trình xây dựng cây RAPTOR bắt đầu bằng việc chuyển đổi các đoạn văn bản thành các vector nhúng ngữ nghĩa sử dụng SBERT. Các vector này sau đó được giảm chiều bằng UMAP để tăng cường hiệu quả phân cụm. Gaussian Mixture Models (GMM) được sử dụng để phân cụm các đoạn văn bản dựa trên độ tương đồng ngữ nghĩa, cho phép mỗi đoạn có thể thuộc về nhiều cụm khác nhau.

\subsubsection{Tóm tắt và Xây dựng Các Nút Cha}

Sau khi phân cụm, mỗi cụm được tóm tắt thành một nút cha bằng GPT-3.5-turbo. Các nút cha này đại diện cho các chủ đề hoặc ý chính của cụm văn bản và trở thành các nút con cho bước phân cụm tiếp theo. Quá trình này lặp lại cho đến khi không thể phân cụm thêm, tạo thành một cây phân cấp với nhiều lớp trừu tượng hóa.

\subsubsection{Chiến lược Truy vấn Hiệu quả}

RAPTOR hỗ trợ hai chiến lược truy vấn chính: Tree Traversal và Collapsed Tree Retrieval. Tree Traversal đi từ lớp gốc đến lớp lá, chọn top-$k$ nút tại mỗi lớp dựa trên độ tương đồng với truy vấn. Collapsed Tree Retrieval gộp toàn bộ cây thành một lớp duy nhất và chọn top-$k$ nút có độ tương đồng cao nhất. Phương pháp này được chứng minh là hiệu quả hơn trong các thử nghiệm trên tập dữ liệu QASPER, nhờ khả năng linh hoạt trong việc chọn các nút ở mức độ trừu tượng phù hợp với yêu cầu truy vấn.

\subsection{Kết luận}

RAPTOR đại diện cho một bước tiến đáng kể trong hệ thống truy xuất thông tin và tạo văn bản, đặc biệt hiệu quả trong các bài toán yêu cầu ngữ cảnh dài hoặc suy luận phức tạp. Với cấu trúc cây phân cấp, phương pháp này không chỉ tăng cường khả năng hiểu ngữ nghĩa mà còn mở rộng phạm vi ứng dụng của các hệ thống truy xuất thông tin. Tuy nhiên, các yêu cầu về tài nguyên tính toán và độ phức tạp triển khai vẫn là những thách thức lớn, đòi hỏi sự cân nhắc kỹ lưỡng khi áp dụng vào thực tế.

\end{document}
```

   - **Chia nhỏ và Nhúng Vector:** Mô tả chi tiết về cách chia nhỏ tài liệu và sử dụng SBERT để nhúng ngữ nghĩa.
   - **Phân Cụm Ngữ Nghĩa:** Giải thích về việc sử dụng Gaussian Mixture Models và UMAP để phân cụm các vector nhúng.
   - **Tóm tắt và Xây dựng Cấu Trúc Cây:** Mô tả quá trình tóm tắt các cụm thành nút cha và xây dựng cây phân cấp đệ quy.

2. **Chiến lược Truy vấn trong Cây RAPTOR:**
   - **Truyền cây theo lớp (Tree Traversal):** Mô tả chi tiết về cách truy vấn lớp theo lớp và gộp kết quả.
   - **Truy vấn cây gộp (Collapsed Tree Retrieval):** Mô tả cách gộp cây thành một lớp duy nhất và tìm kiếm trực tiếp trên toàn bộ cây.

3. **Quá trình Xây dựng và Tối ưu hóa Cấu trúc Cây:**
   - **Embedding và Phân Cụm:** Chi tiết về quá trình nhúng và phân cụm các đoạn văn bản.
   - **Tóm tắt và Xây dựng Các Nút Cha:** Chi tiết về cách tóm tắt các cụm thành nút cha và xây dựng cây phân cấp.
   - **Chiến lược Truy vấn Hiệu quả:** So sánh hiệu quả giữa Tree Traversal và Collapsed Tree Retrieval.

4. **Kết luận:** Tóm tắt lại những điểm mạnh và những thách thức của RAPTOR, đồng thời nhấn mạnh sự cần thiết của việc cân nhắc khi áp dụng phương pháp này vào thực tế.
